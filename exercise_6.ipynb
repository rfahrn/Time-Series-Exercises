### 1. K-means

The K-means algorithm is a clustering algorithm that aims to partition n observations into k clusters. The algorithm works as follows:
- initial cluster centers, given are: cluster center 1 (blue) and cluster center 2 (pink): 
    - cluster center 1: (75, 90)
    - cluster center 2: (65, 45)
- iteration 1: distance calcuation and cluster assignment 
    - calculate eucledian distancees; The time series points are as follows: (85,80), (70,65), (60,50), (40,25), (35,95), (30,70), (25,40), (10,15).
    - for each X = (x1, x2) the euclidean distance to a cluster center C = (c1, c2) is calculated as follows: sqrt((x1-c1)^2 + (x2-c2)^2) so, in first iteration:

        - distance to cluster center 1: sqrt((x1-75)^2 + (x2-90)^2)

        - distance to cluster center 2: sqrt((x1-65)^2 + (x2-45)^2)

        we assign the time series point to the cluster center with the smallest distance.
- iteration 2: update of cluster centers (distance recalculation) 
- sum of squared errors (SSE) calculation for each cluster

usually the algorithm is repeated until the cluster centers do not change anymore.

### 2. Dynamic time wraping (DTW)

Dynamic time warping (DTW) is an algorithm used to compare two time series. It works as follows:
- initial values
- distance matrix calculation (cost matrix)
- optimal path calculation (shortest wraping path)
- distance calculation (DTW distance) between the two time series

This is used to compare different lengths and speeds of time series. In our task we are given these two time series:

Series 1: [3,4,2,3,3,1,4] (m=7)
Series 2: [3,3,4,2,4,3]   (n=6)

in the intial step we initialize the cost matrix with size (m+1) x (n+1) where m is the length of series 1 and n is the length of series 2. The cost matrix is initialized with infinity. The first element of the cost matrix is set to 0.

What we need to calculate is basically: 


1. **Initialize the Cost Matrix:**
$$
C_{i,j} = \infty, \; 	for all \; i, j \ except \;
C_{0,0} = 0
$$

2. **Calculate the cost Matrix:**
$$
D_{i,j} = d(x_i, y_j) + min(D_{i-1,j}, D_{i,j-1}, D_{i-1,j-1})
$$

3. **Calculate the Optimal Path:**
$$
\text{path} = \text{backtrack}(D)
$$

4. **Calculate the DTW Distance:**
$$
\text{DTW}(x, y) = D_{n,m}
$$


```python
import numpy as np
S1 = [3, 3, 4, 2, 4, 3] 
S2 = [3, 4, 2, 3, 3, 1, 4]
S1_labels = ['t0'] + [f't{i}' for i in range(1, len(S1) + 1)]
S2_labels = ['t0'] + [f't{i}' for i in range(1, len(S2) + 1)]

# cost matrix with infinity and set the starting point to 0
D = np.full((len(S1) + 1, len(S2) + 1), np.inf)
D[0, 0] = 0

def print_matrix_with_labels(D, S_labels, S2_labels):
    print('    ' + ' '.join(f'{label:>3}' for label in S2_labels))
    for i, row in enumerate(D):
        print(f'{S_labels[i]:>3} ' + ' '.join(f'{val:>3}' for val in row))
print('initialized matrix:')
print_matrix_with_labels(D, S1_labels, S2_labels)
def shortest_path_cost(D):
    return D[-1, -1]


def calculate_dtw(S1, S2, D, i, j):
    cost = abs(S1[i-1] - S2[j-1])
    D[i, j] = cost + min(D[i-1, j-1], D[i-1, j], D[i, j-1])
    return D

for i in range(1, len(S1) + 1):
    for j in range(1, len(S2) + 1):
        D = calculate_dtw(S1, S2, D, i, j)
        print(f'\n calculation of cost: {abs(S1[i-1] - S2[j-1])} (absolut value of S1(i-1) S2(j-1)) + min({D[i-1, j-1]}, {D[i-1, j]}, {D[i, j-1]}) (directions)')
        print(f'therefore cost at ({S1_labels[i]}, {S2_labels[j]}) is {D[i, j]}')
        print(f'matrix after filling D({S1_labels[i]}, {S2_labels[j]}):\n')
        print_matrix_with_labels(D, S1_labels, S2_labels)
        print()  # Newline for better readability between steps
        
```

final cost matrix looks like this:
 
 $$
    \begin{array}{c|ccccccc}
    & t0 & t1 & t2 & t3 & t4 & t5 & t6 & t7 \\
    \hline
    t0 & 0 & \infty & \infty & \infty & \infty & \infty & \infty & \infty \\
    t1 & \infty & 0 & 1 & 2 & 2 & 2 & 4 & 5 \\
    t2 & \infty & 0 & 1 & 2 & 2 & 2 & 4 & 5 \\
    t3 & \infty & 1 & 0 & 2 & 3 & 3 & 5 & 4 \\
    t4 & \infty & 2 & 2 & 0 & 1 & 2 & 3 & 5 \\
    t5 & \infty & 3 & 2 & 2 & 1 & 2 & 5 & 3 \\
    t6 & \infty & 3 & 3 & 3 & 1 & 1 & 3 & 4 \\
    \end{array}
$$

so the DTW distance between the two series is 4.0, "inf" denotes cells that have not yet been computed or are not reachable based on the constraints of dtw (usually  only movement to the right, down, or diagonally down-right)



```python
def backtrack_optimal_path(D):
    i, j = len(S1), len(S2)
    path = [(i, j)]
    while i > 0 or j > 0:
        directions = []
        if i > 0 and j > 0:
            directions.append((D[i-1, j-1], (i-1, j-1)))
        if i > 0:
            directions.append((D[i-1, j], (i-1, j)))
        if j > 0:
            directions.append((D[i, j-1], (i, j-1)))
        cost, (i, j) = min(directions)
        path.insert(0, (i, j))
    return path

optimal_path_calculated = backtrack_optimal_path(D)
print("\nCalculated optimal path:", optimal_path_calculated)

def backtracking_optimal_path2(D,S1,S2):
    path = []
    i, j = len(S1), len(S2)
    while i > 0 and j > 0:
        path.append((i, j))  
        if i == 0:
            j -= 1
        elif j == 0:
            i -= 1
        else: 
            min_cost = min(D[i-1, j], D[i, j-1], D[i-1, j-1])
            if min_cost == D[i-1, j-1]:
                i -= 1
                j -= 1
            elif min_cost == D[i, j-1]:
                j -= 1
            else:
                i -= 1
    path.append((0, 0))
    return path[::-1]

optimal_path_calculated2 = backtracking_optimal_path2(D,S1,S2)
print("\nCalculated optimal path:", optimal_path_calculated2)

def backtrack_lexicographical(D):
    i, j = len(S1), len(S2)
    path = [(i, j)]
    while (i, j) != (0, 0):
        # If we're at the first row, we can only go left
        if i == 0:
            j -= 1
        # If we're at the first column, we can only go up
        elif j == 0:
            i -= 1
        else:
            # Check the cost for horizontal, diagonal, and vertical moves
            horizontal = D[i, j-1]
            diagonal = D[i-1, j-1]
            vertical = D[i-1, j]
            # Prioritize horizontal move when there's a tie, then diagonal, then vertical
            if horizontal <= diagonal and horizontal <= vertical:
                j -= 1
            elif diagonal < vertical:
                i -= 1
                j -= 1
            else:
                i -= 1
        path.insert(0, (i, j))
    return path

optimal_path_lexicographical = backtrack_lexicographical(D)
print("\nLexicographical optimal path:", optimal_path_lexicographical)
```

or :

        - [(t0,t0), (t1,t1), (t2,t1), (t3,t2), (t4,t3), (t5,t4), (t6,t4), (t6,t5), (t6,t6), (t6,t7)] 

or : 
        - [(t0,t0), (t1,t1), (t2,t1), (t3,t2), (t4,t3), (t5,t4), (t6,t4), (t6,t4), (t6,t5), (t6,6), (t6,7)]
        


Notice, that here we found that according to the backtracking for the optimal path finding we could get different outputs, due to how we choose to break the ties. In the first calculation all potential moves are considered at once and the one with minimum cost is chosen, it there is a tie, "min" will choose the first tuple in list with smallest first element (cost), it cost is same then the second element (direction) is considered.

In the second implementation we chose to prioritize the diagonal path. Since, the diagonal move (i-1.j-1)is checked first, then horizontal, if no tie then minimum cost direction. 


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

S1 = [3, 4, 2, 3, 3, 1, 4]
S2 = [3, 3, 4, 2, 4, 3]

D = np.full((len(S1) + 1, len(S2) + 1), np.inf) # cost matrix initialization - notice inf for all cells
D[0, 0] = 0  # start

# fill accroding to step condition 
for i in range(1, len(S1) + 1):
    for j in range(1, len(S2) + 1):
        cost = abs(S1[i - 1] - S2[j - 1])
        D[i, j] = cost + min(D[i - 1, j], D[i, j - 1], D[i - 1, j - 1]) # accroding to calcuation of lecture slides 

# for highlighting optimal path
mask = np.ones_like(D, dtype=bool)
path1 = backtrack_optimal_path(D)
path2 = backtracking_optimal_path2(D,S1,S2)
path3 = backtrack_lexicographical(D)
for i, j in path1:
    mask[i, j] = False
    
D[D == np.inf] = np.nan # replace inf with nan 
plt.figure(figsize=(10, 8))
sns.heatmap(D, annot=True, fmt=".1f", cmap="viridis", linewidths=.5, mask=mask, cbar=False)
plt.title('DTW Cost Matrix with Optimal Path: backtracking (priority to diagonal)')
plt.xlabel('S2 Series Index')
plt.ylabel('S1 Series Index')
plt.show()
```

The shortest warping path (as defined as minumum cumulative distance) is not unique, but the DTW distance is unique, which is 4.0 in this case. We can implement different backtracking strategies to find the optimal path.
